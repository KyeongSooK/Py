# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호') 
df
```

```python
df['키'] >= 185 # 학생들의 키가 185이상 인지 여부를 True/False 로 볼수 있다
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[-filt] # filter를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
# 키가 185 이상인 학생들의 수학 데이터
df.loc[df['키'] >= 185, '수학'] # df.loc[row_sel, col_sel]
```

```python
# 키가 185 이상인 학생들의 이름, 수학, 과학 데이터
df.loc[df['키'] >= 185, ['이름','수학','과학']]
```

### 다양한 조건

& 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185 이상인 북산고 학생 데이터
```

### 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키가 200보다큰 170보다 작은 데이터
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # 송 씨의 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 태 가 들어가는 사람
df[filt]
```

```python
df[-filt] # 이름에 '태' 가 들어가는 사람은 제외
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # SW특기가 Python 이거나 Jave 인 사람
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java' , na=False)
df[filt]
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터에 대해서 True 로 설정
```

```python
df['SW특기'].str.contains('Java', na=False) # NaN 데이터에 대해서 False 로 설정
```
